### Logistics

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
# Load data from the text file
log_data_file = 'logistics_compute_resultst.txt'  # Replace with your file path
log_data = pd.read_csv(log_data_file, sep='\t')
```

```python
# Filter data to ignore rows where time >= 50000
log_data = log_data[log_data['time'] < 50000]

# Define box plot colors
box_colours = ['#ece2f0', '#1c9099']  # Adjust if you have more hues
# box_colours = ['#edf8b1', '#7fcdbb', '##2c7fb8']  # Adjust if you have more hues
# box_colours = ['#e0ecf4', '#8856a7']  # Adjust if you have more hues

# Sort the unique number of robots
unique_num_rob = log_data['num_rob'].unique()
unique_num_rob.sort()

# Set up the figure
plt.figure(figsize=(7, 6))

# Customize flier properties for small anomalous points
flierprops = dict(marker='o', color='red', markersize=1)

# Create a combined plot
sns.boxplot(
    data=log_data,
    x='num_rob',
    y='compu_time_comms',
    hue='type',  # Differentiate by type (e.g., naive, optimized)
    palette=box_colours,
    notch=True,
    width=0.6,
    flierprops=flierprops,
    order=unique_num_rob,  # Ensure consistent order of robot numbers
    hue_order=['naive', 'optimised']  # Explicit order of 'type'
)

# Remove gridlines
plt.grid(False)

# Add titles and labels
plt.title('(a) - Logistics Task', fontsize=14)
plt.xlabel('Number of Robots', fontsize=14)
plt.ylabel('Ave. compute time for communications per tick, s', fontsize=14)

# Adjust legend appearance
plt.legend(title='Type', title_fontsize=14, fontsize=14)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Logistics

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
# Load data from the text file
log_data_file = 'logistics_results.txt'  # Replace with your file path
log_data = pd.read_csv(log_data_file, sep='\t')
```

```python
# Filter data to ignore all rows where time >= 50000
log_data = log_data[log_data['time'] < 50000]

# Check the data
print(log_data.head())

#box_colors = ['#ece2f0', '#a6bddb', '#1c9099']  # Adjust if you have more hues
box_colours = ['#edf8b1', '#7fcdbb', '#2c7fb8']  # Adjust if you have more hues
# box_colors = ['#e0ecf4', '#9ebcda', '#8856a7']  # Adjust if you have more hues

# Create individual box plots for each num_rob
unique_num_rob = log_data['num_rob'].unique()
unique_num_rob.sort()

# Set up the figure with subplots
plt.figure(figsize=(10, 5))

for i, num_rob in enumerate(unique_num_rob, 1):
    plt.subplot(1, len(unique_num_rob), i)
    subset = log_data[log_data['num_rob'] == num_rob]
    
    # Customize flier properties for small anomalous points
    flierprops = dict(marker='o', color='red', markersize=1)
    
    # Use seaborn boxplot with a colorblind-friendly pastel palette
    sns.boxplot(
        data=subset,
        x='type',
        y='time',
        palette=box_colours,
        notch=True,
        width=0.7,
        flierprops=flierprops,
        order = ['baseline', 'naive', 'optimised']
    )
    
    # Remove gridlines
    plt.grid(False)
    
    
    
    plt.title(f'Robots: {num_rob}', fontsize=12)
    plt.xlabel('Type', fontsize=12)
    plt.ylabel('Timesteps', fontsize=12) if i == 1 else plt.ylabel('')

# Adjust layout for horizontal compression
plt.tight_layout()
plt.show()
```

### Traffic Results

```python
# Load data from the text file
tr_data_file = 'traffic_results.txt'  # Replace with your file path
tr_data = pd.read_csv(tr_data_file, sep='\t')
```

```python
# Filter data to ignore rows where time >= 50000
tr_data = tr_data[tr_data['time'] < 50000]

# Define box plot colors
box_colours = ['#edf8b1', '#7fcdbb', '#2c7fb8']

# Sort the unique number of robots
unique_num_rob = tr_data['num_rob'].unique()
unique_num_rob.sort()

# Set up the figure with subplots
plt.figure(figsize=(10, 5))

for i, num_rob in enumerate(unique_num_rob, 1):
    plt.subplot(1, len(unique_num_rob), i)
    subset = tr_data[tr_data['num_rob'] == num_rob]
    
    # Customize flier properties for small anomalous points
    flierprops = dict(marker='o', color='red', markersize=1)
    
    # Use seaborn boxplot with a colorblind-friendly pastel palette
    sns.boxplot(
        data=subset,
        x='type',
        y='time',
        palette=box_colours,
        notch=True,
        width=0.6,
        flierprops=flierprops,
        order=['baseline', 'naive', 'optimised']
    )
    
    # Set y-axis limit for specific num_rob values
    if num_rob in [100, 200]:
        plt.ylim(1000, 4000)
    
    # Remove gridlines
    plt.grid(False)
    
    # Add titles and labels
    plt.title(f'Robots: {num_rob}', fontsize=12)
    plt.xlabel('', fontsize=12)
    plt.ylabel('Timesteps', fontsize=12) if i == 1 else plt.ylabel('')

# Adjust layout for horizontal compression
plt.tight_layout()
plt.show()
```

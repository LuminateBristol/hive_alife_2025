### Logistics

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data from the text file
log_data_file = 'logistics_results.txt'  # Replace with your file path
log_data = pd.read_csv(log_data_file, sep='\t')
```

```python
# Filter data to ignore all rows where time >= 50000
log_data = log_data[log_data['time'] < 50000]

# Check the data
print(log_data.head())

# Create individual box plots for each num_rob
unique_num_rob = log_data['num_rob'].unique()
unique_num_rob.sort()

# Set up the figure with subplots
plt.figure(figsize=(15, 6))

for i, num_rob in enumerate(unique_num_rob, 1):
    plt.subplot(1, len(unique_num_rob), i)
    subset = log_data[log_data['num_rob'] == num_rob]
    
    # Use seaborn boxplot with a colorblind-friendly pastel palette
    sns.boxplot(
        data=subset,
        x='type',
        y='time',
        palette=sns.color_palette("pastel"),
        notch=True,
        width=0.6
    )
    
    plt.title(f'num_rob = {num_rob}', fontsize=12)
    plt.xlabel('Type', fontsize=10)
    plt.ylabel('Time', fontsize=10) if i == 1 else plt.ylabel('')

# Adjust layout for horizontal compression
plt.tight_layout()
plt.show()
```
